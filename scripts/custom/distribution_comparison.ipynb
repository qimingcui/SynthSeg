```python
output_labels = np.load('priors/output_labels.npy',allow_pickle=True)
output_labels = np.insert(output_labels, 0, 0)
generation_labels = np.load('priors/generation_labels.npy',allow_pickle=True)
generation_labels = np.insert(generation_labels, 0, 0)
classes_names = np.load('priors/classes_names.npy',allow_pickle=True)
generation_classes = np.load('priors/generation_classes.npy',allow_pickle=True)
generation_classes = np.insert(generation_classes, 0, 0)
t1_mean = np.load('priors/t1_mean_class_based.npy')
# t1_mean = np.insert(t1_mean, 0, [0,0], axis=1)
t1_std = np.load('priors/t1_std_class_based.npy')
# t1_std = np.insert(t1_std, 0, [0,0], axis=1)
t2_mean = np.load('priors/t2_mean_class_based.npy')
# t2_mean = np.insert(t2_mean, 0, [0,0], axis=1)
t2_std = np.load('priors/t2_std_class_based.npy')
# t2_std = np.insert(t2_std, 0, [0,0], axis=1)
pd_mean = np.load('priors/pd_mean_class_based.npy')
# pd_mean = np.insert(pd_mean, 0, [0,0], axis=1)
pd_std = np.load('priors/pd_std_class_based.npy')
```

```python
import matplotlib.pyplot as plt
from scipy.stats import norm

# Create figure with 3 subplots
fig, (ax1, ax2, ax3) = plt.subplots(3, 1, figsize=(12, 6))

# Function to plot gaussian distributions
def plot_gaussian(ax, mean, std, title):
    # Calculate appropriate x range for each gaussian separately
    x1 = np.linspace(mean[0] - 4*std[0], mean[0] + 4*std[0], 1000)
    x2 = np.linspace(mean[1] - 4*std[1], mean[1] + 4*std[1], 1000)
    
    # Plot upper and lower bound gaussians
    ax.plot(x1, norm.pdf(x1, mean[0], std[0]))
    ax.plot(x2, norm.pdf(x2, mean[1], std[1]))
    
    ax.set_title(title)
    ax.legend()
    ax.grid(True)

# Plot T1 distributions
plot_gaussian(ax1, t1_mean, t1_std, 'T1 Distribution')
ax1.set_xlabel('T1 value')
ax1.set_ylabel('Probability Density')

# Plot T2 distributions
plot_gaussian(ax2, t2_mean, t2_std, 'T2 Distribution')
ax2.set_xlabel('T2 value')
ax2.set_ylabel('Probability Density')

# Plot PD distributions
plot_gaussian(ax3, pd_mean, pd_std, 'PD Distribution')
ax3.set_xlabel('PD value')
ax3.set_ylabel('Probability Density')

plt.tight_layout()
plt.show()
```
